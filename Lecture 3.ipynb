Построим свой алгоритм линейной регрессии, на 2-х зависымых переменнных
Сначала мы построим данные для обучения: возьмем прямую линию и наложим на нее нормального шума

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -0.5 + 1.3*x
```

```python
x = np.arange(-5,5,.1)
M = len(x)
noise = np.random.normal(0,2,M)
y = f(x) + noise
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x, y, color='tab:blue')
ax.plot(x, f(x), color='tab:orange')
```

Посмотрим на ошибку

```python
J = sum((x-f(x))**2)/M
J
```

Напишем свой простой алгоритм линейной регрессии

```python
Theta_0 = 0
Theta_1 = 0
rate = 0.1
```

```python
def draw(x,h,f):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.scatter(x, y, color='tab:blue')
    ax.plot(x, h, color='tab:orange')
    ax.plot(x, f(x), color='tab:red')  
```

```python
for s in range(20):
    h = Theta_0 + x*Theta_1  #?    
    J = ((h-y)**2).sum()/M   #?
    
    print("Step : ", s, ", Error: ",J, ", Theta_0: ", Theta_0, ", Theta_1: ", Theta_1)
    draw(x,h,f)
    
    DJ_Theta_0 = sum(h-y)/M   # ?
    DJ_Theta_1 = sum(np.multiply((h-y),x))/M # ?
    
    Theta_0 =  Theta_0 -rate * DJ_Theta_0
    Theta_1 =  Theta_1 -rate * DJ_Theta_1
    
    print("DTheta0 = ", DJ_Theta_0)
    print("DTheta1 = ", DJ_Theta_1)
    
    h = Theta_0 + x*Theta_1

draw(x,h,f)
    
```
